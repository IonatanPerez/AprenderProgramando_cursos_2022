```python
#Punto 3

grupo_1 = ["Agustina", "Cristian", "Tomas"]
grupo_2 = ["Gabriel", "Samuel", "Delfina"]
grupo_3 = []

for i in range(len(grupo_1)):
  grupo_3.append(grupo_1[i])

grupo_1.clear

for i in range(len(grupo_2)):
  grupo_3.append(grupo_2[i])

grupo_2.clear

print("Grupo: " + str(grupo_3))
```

```python
#Punto 4

todos = []


grupo_1 = ["Agustina", "Cristian", "Tomas"]
grupo_2 = ["Gabriel", "Samuel", "Delfina"]
grupo_3 = []

for i in range(len(grupo_1)):
  grupo_3.append(grupo_1[i])
  todos.append(grupo_1[i])

for i in range(len(grupo_2)):
  grupo_3.append(grupo_2[i])
  todos.append(grupo_2[i])

for i in range(len(grupo_3)):
  todos.append(grupo_3[i])

grupo_1.clear
grupo_2.clear
grupo_3.clear

print("Grupo: " + str(todos))
```
