```python
#Armar una lista con 6 ciudades
ciudades=["lima","capital","cordoba","chaco","tierra del fuego","salta"]
print(ciudades)
#Seleccionar la segunda ciudad
segunda_ciudad=ciudades[1:2]
print(segunda_ciudad)
#Eliminar la última ciudad
ciudades.pop()
print(ciudades)
#Reemplazar la primer ciudad con otra
ciudades[0]="misiones"
print(ciudades)
#Agregar una ciudad al final
ciudades.append("caracas")
print(ciudades)
#Armar un diccionario que tenga 3 duplas de Provincia y Ciudad (ej: "Provincia de Buenos Aires: La Plata"
diccionario={"Entre Ríos": "Paraná","Tucumán":"San Miguel de Tucumán","Río Negro":"Viedma"}
print(diccionario)
#Eliminar el último registro (dupla Ciudad-Provincia)
diccionario.popitem()
print(diccionario)
#Ver todas las claves (keys)
print(diccionario.keys())


```
