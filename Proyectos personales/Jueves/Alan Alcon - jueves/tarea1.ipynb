```python
def cheq (nombre,lista1,lista2):
  if nombre in lista1:
    if nombre in lista2:
      print (nombre, "esta inscripto")
    else:
      print (nombre, "no esta inscripto!")

lista_de_presentes = ["Camila 1","Breider","Alan","Joaquin","Sergio"]
lista_de_inscriptes = ["Camila 1","Camila 2","Breider","Alan","Ayax","Joaquin","Marco"]

for nombre in lista_de_presentes:
  cheq (nombre,lista_de_presentes,lista_de_inscriptes)
```
