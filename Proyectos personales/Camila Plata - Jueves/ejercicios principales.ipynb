Ejericicios estructuras de control

```python
#1.definir 2 variables con números y crear una estructura condicional donde si los dos números son iguales se imprima "los dos números son iguales":

num1= 5
num2= 5
if num1==num2:
  print("son iguales")
else:
  print("son distintos")
```

```python
#2.crar una estructura condicional donde si "no llueve", "podemos ir a la plaza" y sino "debemos quedarnos en casa":

no_llueve= 1
llueve= 1
if no_llueve==llueve:
  print("podemos ir a la plaza")
else:
  print("debemos quedarnos en casa")
```

```python
#3.definir una lista con 3 numeros y crear una estructura iterativa donde se imprima cada numero +2 (for..in):

lista_numeros= [3, 7, 15]
for i in lista_numeros:
  print(i+2)
```

```python
#4.crear una lista con 5 materias y una estructura iterativa que permita contar la cantidad de elementos en la lista(for..in):

lista_materias= ["civica", "fisica", "lengua", "proyecto", "algoritmos"]
for i in lista_materias:
  print(i)
```

```python
#5.crear una funcion que divida 2 numeros que reciba como parametros:
def funcion_numeros(num1, num2):
  return num1/num2
funcion_numeros(350,50)
```

```python
#6.crear una funcion que reciba 2 numeros e incluya un condicional donde si el num2 es 0 se imprima "no se puede realizar esta operación" y sino realice una división entre num1 y num 2:

def funcion_num(num3, num4):
  if num3==0:
    print("No se puede realizar esta operacion")
  else:
    return num3/num4
funcion_num(0,50)
```
