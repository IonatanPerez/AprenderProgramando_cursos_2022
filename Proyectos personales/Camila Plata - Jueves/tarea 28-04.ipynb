```python
# Vamos a definir una funcion que le paso el nombre y la lista de presentes y me dice si esta presente o no
def chequear_presente(nombre_de_persona,lista_de_curso):
  if nombre_de_persona in lista_de_curso:
    print (nombre_de_persona, "esta presente.")
  else:
    print (nombre_de_persona, "esta ausente.")

# Cargar datos

lista_de_presentes = ["Camila 1","Breider","Alan","Joaquin","Sergio"]

lista_de_inscriptes = ["Camila 1","Camila 2","Breider","Ayax","Alan","Joaquin","Marco"]

# Ejecutamos la parte principal

for nombre in lista_de_inscriptes:
  chequear_presente(nombre,lista_de_presentes)


# Queda de tarea que hagan una version que chequee les infiltrades
```

```python
#chequear a los infiltrados
def chequear_infiltrados(nombre_de_persona,lista_de_curso):
  if nombre_de_persona in lista_de_curso:
    print (nombre_de_persona, "no infiltrado.")
  else:
    print (nombre_de_persona, "¡ALERTA!, infiltrado.")

lista_de_presentes = ["Camila 1","Breider","Alan","Joaquin","Sergio"]

lista_de_inscriptes = ["Camila 1","Camila 2","Breider","Ayax","Alan","Joaquin","Marco"]

for nombre in lista_de_presentes:
  chequear_infiltrados(nombre,lista_de_inscriptes)
```
