```python
import pandas as pd
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
data = pd.read_csv("pokemon.csv")
data.head()
```

```python
data.drop(columns=['nombre','indice_guia'], inplace=True)
```

```python
nombres = {'tipo_primario' : 'primario', 'tipo_secundario' : 'secundario', 'Weight' : 'peso', 'Height' : 'altura', 'Legendario' : 'legendario'}
data.rename(columns = nombres, inplace=True)
data.head()
```

```python
data.isna().sum()
```

```python
filtro = data['ataque'].isna()
data[filtro]
data.dropna(thresh=6, inplace=True)
```

```python
filtro = data['altura'].isna()
promedio = data[~filtro]['altura'].mean()
data.loc[filtro, 'altura'] = promedio
data[filtro]
```

```python
data.isna().sum()
filtro = data['legendario'].isna()
data[filtro]
data['legendario'].value_counts()
data.loc[filtro,'legendario'] = 'No'
data.isna().sum()
```

```python
data['primario'].value_counts()
pd.get_dummies(data)
```

```python
data = pd.get_dummies(data, drop_first = True)
data.head()
```

```python
y = data['fuerza_combate']
X = data.drop(columns = 'fuerza_combate') 
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=55)
```

```python
from sklearn.linear_model import LinearRegression 
mostrar = LinearRegression()
mostrar.fit(X_train,y_train)

print (mostrar.coef_)
```

```python
print (X_test.iloc[0])
print (y_test.iloc[0])
mostrar.predict([X_test.iloc[0]])
```

```python
from sklearn.metrics import r2_score

ypred = mostrar.predict(X_test)
y_test
```
