```python
# Ejercicio Nro: 2

grupo1 = ["Facundo", "Sammy", "Alex"]
grupo2 = ["Rodrigo", "Alvarez", "Lucio"]

print("Lista nro. 1 y nro. 2: ")
print(grupo1, grupo2)

# Ejercicio Nro: 3

nombre1 = grupo1.pop()
nombre2 = grupo2.pop()
grupo3 = [nombre1, nombre2]

print("Lista nro. 3: ")
print(grupo3)

# Ejercicio Nro. 4

todos = [grupo1 + grupo2 + grupo3]
print("Lista nro. 4: ")
print(todos)
```
