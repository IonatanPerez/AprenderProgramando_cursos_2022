```python
def suma(primer_numero,segundo_numero):
  if (not type(primer_numero) == "int") or (not type(primer_numero) == "float"):
    print ("Error: Tipo de datos no numerico!")
    return None
  if (not type(segundo_numero) == "int") or (not type(segundo_numero) == "float"):
    print ("Error: Tipo de datos no numerico!")
    return None
  return primer_numero + segundo_numero
```

```python
suma("Hola","chau")
```

```python
import seaborn as sns
```

```python
data = sns.load_dataset("penguins")
data.head()
```

```python
sns.countplot(data["species"])
```

```python
data_agrupada = data.groupby("species")
data_agrupada["body_mass_g"].mean()
```

```python
sns.barplot(x = data_agrupada["body_mass_g"].min().index, y=data_agrupada["body_mass_g"].min() )
plt.xticks(rotation=180)
plt.yticks(rotation=90)
```

```python
sns.displot(data=data, x="bill_length_mm", hue="species")
```
