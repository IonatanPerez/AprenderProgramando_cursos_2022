```python
def cambiar_valor(valor):
  if valor == 'si':
    return 1
  else:
    return 0
```

```python
# Estamos chequeando que hay en lluvia
sns.countplot(data['lluvia'])
data['lluvia'].unique()
```

```python
data['lluvia_binarizada'] = data['lluvia'].apply(cambiar_valor)

data.head()
```

```python
data['lluvia'].apply(lambda x: 1 if x=='si' else 0)

dict_lluvia = {'si':1, 'no':0}

data['lluvia'].map(dict_lluvia)
```

```python
# Vamos a ver como procesar la columna descripcion que tiene varias categorias. No podemos pasarlo todo a una sola columna de numeros
display (data['descripcion'].head())
pd.get_dummies(data['descripcion'], drop_first=True)

data_limpia = data.join(pd.get_dummies(data['descripcion'], drop_first=True, prefix='descripcion'))

data_limpia.head()
data_limpia.drop(columns = ['lluvia','descripcion'], inplace=True)

data_limpia.head()

# data_limpia.drop(columns = 'Cold')
```
