```python
data['encuentro_participantes'].sum()
```

# Consigna

Queremos saber cuantos asistentes hubo a los encuentros en un año en particular.

```python
ano_seleccionado = 2017
mascara_seleccion_ano = data['encuentro_anio'] == ano_seleccionado
print (mascara_seleccion_ano)
```

```python
data[mascara_seleccion_ano]["encuentro_participantes"].sum()
```

Ahora queremos repetirlo para todos los años

```python
# Queremos conocer la lista de años que tenemos que mirar.

lista_de_anos = data['encuentro_anio'].unique()

for ano in lista_de_anos:
  mascara_seleccion_ano = data['encuentro_anio'] == ano
  cantidad_en_ano = data[mascara_seleccion_ano]["encuentro_participantes"].sum()
  print (f"La cantidad de participantes que asistieron en {ano} fue de {cantidad_en_ano}")
```

```python
data_agrupada = data.groupby("encuentro_anio")
data_agrupada["encuentro_participantes"].sum()
```

```python
# Queremos saber cuanto encuentros hubo en cada año

data_agrupada["encuentro_anio"].count()
```

```python
# Queremos saber el promedio de participantes en cada reunion cada año

data_agrupada["encuentro_participantes"].mean()
```

```python
# Queremos saber en cada año cual fue la reunion mas chica
data_agrupada["encuentro_participantes"].min()
data_agrupada["encuentro_participantes"].max()
```
