Ahora que ya entendimos como buscar los presentes, vamos a volver a escribir estoc omo una funcion linda y separada

```python
def buscar_presentes(inscriptos_del_curso,presentes_del_dia,fecha):
  print ("Lista de presentes del dia:", fecha)
  for nombre in inscriptos_del_curso:
    esta_presente = nombre in presentes_del_dia
    if esta_presente == False:
      print (nombre, "falto a clase")
    else:
      print (nombre, "esta presente")
```

```python
buscar_presentes(lista_de_inscriptos_curso_martes,lista_de_presentes_26_04_22,"26 de abril")
```
